# An interactive form for details


```python
# In this we have first asked for four details 
# After four details have been given
# we ask if something is wrong in what the voice_recognition has perceived
# Then we ask which one is wrong?
# We correct the given promt till it is correct
# and finally if any second person wants to aslo add data
# if not data is printed 
# Some other small issues have also been addressed : --like if not understood

import pyttsx3
import speech_recognition as sr
import pandas as pd


def speak(text):
    engine = pyttsx3.init()
    engine.say(text)
    engine.runAndWait()

def listen(question):
    recognizer = sr.Recognizer()
    with sr.Microphone() as source:
        print(question)
        speak(question)
        recognizer.adjust_for_ambient_noise(source)
        audio = recognizer.listen(source)

    try:
        text = recognizer.recognize_google(audio)
        print("You said:", text)
        return text
    except sr.UnknownValueError:
        print("Sorry, I could not understand your response.")
        speak("Sorry, I could not understand your response.")
        return listen(question)  

def fill_form():
    name = listen("What is your name?")
    roll_no = listen("What is your roll number?")
    institution = listen("Which institution do you belong to?")
    state = listen("Which state are you from?")

    return name, roll_no, institution, state

def main():
    df = pd.DataFrame(columns=['Name', 'Roll No', 'Institution', 'State'])

    speak("Please provide the following details:")
    pprint("Please provide the following details: ")
    name, roll_no, institution, state = fill_form()

    if name is not None and roll_no is not None and institution is not None and state is not None:
        df = df.append({'Name': name, 'Roll No': roll_no, 'Institution': institution, 'State': state}, ignore_index=True)

    while True:
        print("All data recorded:")
        print(df)
        
        correction = listen("Review the recorded data. If something is wrong, please say 'yes'. If everything is correct, please say 'no'.")
        if correction and 'yes' in correction.lower():
            while True:
                incorrect_info = listen("Which information is incorrect? Please say 'name', 'roll number', 'institution', or 'state'.")
                if incorrect_info == 'name':
                    name = listen("Please provide your name again.")
                    df.at[df.index[-1], 'Name'] = name
                elif incorrect_info == 'roll number':
                    roll_no = listen("Please provide your roll number again.")
                    df.at[df.index[-1], 'Roll No'] = roll_no
                elif incorrect_info == 'institution':
                    institution = listen("Please provide your institution again.")
                    df.at[df.index[-1], 'Institution'] = institution
                elif incorrect_info == 'state':
                    state = listen("Please provide your state again.")
                    df.at[df.index[-1], 'State'] = state
                else:
                    speak("Sorry, I didn't understand. Please specify what information is incorrect.")
                    continue
                
                confirmation = listen("Is all information correct now?")
                if confirmation and 'no' in confirmation.lower():
                    continue  
                else:
                    break 
            print("All data recorded after correction:")
            print(df)

        response = listen("Do you want to enter details for another person?")
        if response is not None and 'yes' not in response.lower():
            break

        speak("Please provide the following details for the next person:")
        pprint("Please provide the following details for the next person: ")
        name, roll_no, institution, state = fill_form()

        if name is not None and roll_no is not None and institution is not None and state is not None:
            df = df.append({'Name': name, 'Roll No': roll_no, 'Institution': institution, 'State': state}, ignore_index=True)

    print("Final recorded data:")
    print(df)
    speak("All data recorded. Thank you!")

if __name__ == "__main__":
    main()
```
